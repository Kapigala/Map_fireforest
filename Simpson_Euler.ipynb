```python
import numpy as np
import itertools  
import pandas as pd
full_values = [(5248, 640),
                   (1312, 1888),
                   (2624, 3776),
                   (5760, 3776),
                   (3936, 5664)] #list of elements
A=[el[0] for el in full_values]
print(A)
B=[el[1] for el in full_values]
print(B)
print("Suma A:",sum([el[0] for el in full_values]),"Suma B:",sum([el[1] for el in full_values]))
```

```python
def m_value(A,B,i):
    x1=[a for a in range(1,A[i]+1)]   #all possible x (errors in "A")
    y1=[b for b in range(1,B[i]+1)] #all possible y (errors in "B")
    c = list(itertools.product(x1, y1))
    df_t=pd.DataFrame(c,columns=["Xi","Yi"])  
    df_t["m"]=np.array([A[i]/B[i]*v[1]/v[0] for v in c ]) #all pairs x,y  and m ratio 
    return df_t
```

```python
# Valid pair of xi,yi should satisfy condition:
#xi*m/Ai = yi/Bi => xi*m*Bi/Ai=yi //m>1  -> yi > xi*Bi/Ai (**)
#what is more  295m|246*X and 246|295*m*Y
```

```python
# 5 groups of xi,yi pairs where m>1 
ll=[m_value(A,B,i) for i in range(0,5)]
for i in range(0,len(ll)):
    ll[i]=ll[i][ll[i]["m"]>1476/1475] # m>1 and m>=1476/1475 according to Project_Euler 
#//answer >1476/1475
```

```python
# consired only m appearing in all 5 groups
m_candidates=ll[0][(ll[0]["m"].isin(ll[1]["m"].unique())) & (ll[0]["m"].isin(ll[2]["m"].unique())) & (ll[0]["m"].isin(ll[3]["m"].unique()) & (ll[0]["m"].isin(ll[4]["m"].unique())))]["m"]
```

```python
# after some precomputations we could set condition m < 12.55
m_candidates = np.unique(m_candidates[m_candidates<12.55])
for i in range(0,len(ll)):
    ll[i]=ll[i][ll[i]["m"].isin(m_candidates)]
```

```python
#redukcja mało licznych wariantów
test_big=[]
for m in sorted(m_candidates,reverse=True):  #możliwe m we wszsytkich
    #if m=1476/1475
    print(f"TEST dla {m}")
    
    aa= ll[0][(ll[0]["m"]==m)]["Yi"].unique()
    bb= ll[1][(ll[1]["m"]==m)]["Yi"].unique()
    cc= ll[2][(ll[2]["m"]==m)]["Yi"].unique()
    dd= ll[3][(ll[3]["m"]==m)]["Yi"].unique()
    ee= ll[4][(ll[4]["m"]==m)]["Yi"].unique()
    if len(aa)*len(bb)*len(cc)*len(dd)*len(ee) >10**6:
        test_big.append(m)
        print(f'Skipping... {m}\n')
        continue
    test_value= list(itertools.product(aa,bb,cc,dd,ee))
    print(len(test_value))
    for k in test_value:
        #krotka y
        if ((np.dot(np.array(list(k)),(np.array(A)/np.array(B)))/sum(list(k)))**0.5 == m):
            print(f"{m} for set {k}")
            ANSWER_1 = (m,k)
            break   
```
